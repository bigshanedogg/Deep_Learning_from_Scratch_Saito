## Implementing activation function
1) step_function<br>
2) sigmoid_function<br>
3) ReLU_function<br>
4) identity_function<br>
5) softmax_function<br>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5.0, 5.0, 0.5)
```

```python
#1) step_function : 0보다 크면 1, 0보다 작으면 0을 반환한다. 잘 사용하지 않는다.
def step_function(x) : 
    y = x>0
    return y.astype(np.int)
```

```python
plt.figure()
plt.plot(x, step_function(x))
plt.title("step_function")
plt.show()
```

```python
#2) sigmoid_function : 주어진 값을 0~1 사이의 값으로 변환해주는 S자 모양의 함수. 0보다 크면 0.5보다 크게, 0보다 작으면 0.5보다 작게 변환된다.
def sigmoid_function(x) : 
    return (1/(1+np.exp(-x)))
```

```python
plt.figure()
plt.plot(x, sigmoid_function(x))
plt.title("sigmoid_function")
plt.show()
```

```python
#3) ReLU : sigmoid가 갖고 있는 큰 값이 충분히 반영되지 않고 1로 수렴하는 문제를 해결하는 활성화 함수
def ReLU_function(x) :
    return np.maximum(0, x)
```

```python
plt.figure()
plt.plot(x, ReLU_function(x))
plt.title("ReLU_function")
plt.show()
```

```python
#4) identity_function : 마지막 출력 함수에서 주로 사용하는 활성화 함수. 형식을 맞추기 위해 사용될 뿐 실제 의미는 없다.
def identity_function(x) :
    return x
```

```python
plt.figure()
plt.plot(x, identity_function(x))
plt.title("identity_function")
plt.show()
```

```python
#5) softmax_function
def softmax_function(x) : 
    return np.exp(x-max(x))/np.sum(np.exp(x-max(x)))
```

```python
plt.figure()
plt.plot(x, softmax_function(x))
plt.title("softmax_function")
plt.show()
```

## 기타 개념
- batch size : 1 epoch에 몇 개의 데이터 행을 학습할 것인지, 그 길이(개수)를 batch size라 한다. 
    - cf) mini-batch, full-batch
- epoch : 1회 batch size씩 나눠서 전체 행을 학습하기 위해 반복해야하는 횟수를 epoch라 한다.
- cross validation : 교차 검증, 데이터셋에서 여러 개의 다른 훈련/시험 집합을 추출해 학습 parameter를 조정하고 비교적 범용적 성능을 평가하기 위한 방법 
    - cf) k-fold CV, leave-one-out, leave-p-out
